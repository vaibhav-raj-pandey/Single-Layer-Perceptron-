```python
from sklearn.datasets import make_moons
```

```python
import numpy as np 
import matplotlib.pyplot as plt
#X, y = make_blobs(n_samples=500, centers=2, n_features=2,random_state=10)
X,y = make_moons(n_samples=1000, shuffle=True, noise=0.2, random_state=1)
```

```python
print(X.shape,y.shape)
plt.style.use('seaborn')
plt.scatter(X[:,0],X[:,1],c=y, cmap= plt.cm.Accent)
```

```python
def sigmoid(Z):
    return (1.0)/(1+np.exp(-Z))
    
```

```python
def predict(X,weights):
    z= np.dot(X,weights)
    predict=sigmoid(z)
    return predict
def loss(X,Y,weights):
    Y_=predict(X,weights)
    loss= np.mean(-Y*np.log(Y_)-(1-Y)*np.log(1-Y_))
    return loss
def update(X,Y,weights,learning_rate):
    Y_=predict(X,weights)
    dw=np.dot(X.T,Y_-y)
    m=X.shape[0]
    weights= weights-learning_rate*dw/(float(m))
    return weights
```

```python
def train(X,Y,learning_rate=0.6,no_of_epochs=1000):
    ones= np.ones((X.shape[0],1))
    X=np.hstack((ones,X))
    weights=np.zeros(X.shape[1])

    for epoch in range(no_of_epochs):
        weights= update(X,Y,weights,learning_rate)
       
        if epoch%5==0:
            l=loss(X,Y,weights)
            print((epoch,l))
    return weights
            
  
                            
```

```python
weights=train(X,y,no_of_epochs=1000)
```

```python
x1 = np.linspace(-2,2,10)
x2 = -(weights[0] + weights[1]*x1)/weights[2]
plt.scatter(X[:,0],X[:,1],c=y, cmap= plt.cm.Accent)
plt.plot(x1,x2,c='purple')
plt.show()
```

```python
def getPredictions(X_Test,weights,labels=True):
    if X_Test.shape[1] != weights.shape[0]:
        ones = np.ones((X_Test.shape[0],1))
        X_Test = np.hstack((ones,X_Test))

    probs = predict(X_Test,weights)
    
    if not labels:
        return probs
    else:
        labels = np.zeros(probs.shape)
        labels[probs>=0.5] = 1
        return labels
Y_ = getPredictions(X,weights,labels=True)
training_acc = np.sum(Y_==y)/y.shape[0]
print(training_acc)
```
